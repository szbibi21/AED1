```python
def valida_cpf (cpf):
    
    # Validação do digito
    
    aux = 10
    soma = 0
    for i in range(0, 11):
        if i == 3 or i == 7: 
            continue
        digito_atual = int(cpf[i])
        soma += digito_atual * aux
        aux -= 1
    resto = soma % 11
    if resto < 2:
        digito_verificador = 0
    else:
        digito_verificador = 11 - resto 

    aux2 = 11
    soma = 0

    for i in range (0,13):
        if i == 3 or i == 7 or i == 11:
            continue
        digito_atual = int(cpf[i])
        soma += digito_atual * aux2
        aux2 -= 1

    resto = soma % 11
    if resto < 2:
        digito_verificador2 = 0
    else:
        digito_verificador2 = 11 - resto

    if digito_verificador == int(cpf[12]) and digito_verificador2 == int(cpf[13]):
        return True
    else:
        return False 

valida_cpf("490.900.288-03")
    
```

```python
from datetime import datetime
def valida_data (data):

    # Verificar se a pessoa é maior de idade

    hoje = datetime.now().date()

    # Separar dia, mês e ano

    dia = int(data[0:2])
    mes = int(data[3:5])
    ano = int(data[6:10])

    data_nasc = datetime(ano, mes, dia).date()
    idade = hoje.year - data_nasc.year - ((hoje.month, hoje.day) < (data_nasc.month, data_nasc.day))

    if idade < 18:
        return False
    else:
        return True

valida_data("30/07/2000")
```

```python
from random import choice
 
def exibir_mensagem():
    mensagens = ["A persistência realiza o impossível", "Seus sonhos não precisam de plateia, eles só precisam de você",
                "A persistência é o caminho do êxito", "No meio da dificuldade encontra-se a oportunidade"]
    return choice(mensagens)
```

```python
def cadastrar():
    while True:
        nome = input("Digite o seu primeiro nome")
        sobrenome = input("Digite o seu sobrenome")

        while True:
            cpf = input("Indique o seu cpf")
            if valida_cpf(cpf):
                break
            else:
                print("Cpf inválido, digite novamente")
                continue
        while True:
            data_nasc = input("Digite a sua data de nascimento")
            if valida_data(data_nasc):
                break
            else:
                print("Menor de idade")
                continue
        
        renda_bruta = float(input("Digite a sua renda bruta"))

        print('\n' + 'Dados Cadastrados:'.center(100))
        print(f'''
        Nome: {nome}
        Sobrenome: {sobrenome}
        CPF: {cpf}
        Data Nascimento: {data_nasc}
        Renda Bruta: {renda_bruta}''')
        print('----- Cadastro bem sucedido! -----'.center(100) + '\n')
        print('Retornando ao menu principal. . .')
        break
```

```python
def exibir_menu():
    while True:
        print('''\nMenu:
        1 - Cadastrar
        2 - Exibir Frase
        3 - Sair''')
        opcao = int(input('\nDigite a opção desejada'))
 
        if opcao == 1:
            cadastrar()
            continue
        elif opcao == 2:
            print('\n' + exibir_mensagem() + '\n')
            continue
        elif opcao == 3:
            print('Bye bye!')
            break
        else:
            print('Opção inválida! Tente novamente.')
            continue
```
