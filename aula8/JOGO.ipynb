```python
import random

print("=== Jogo de Adivinhação ===")
print("Vou pensar em um número de 1 a 50. Tente adivinhar!")

numero_secreto = random.randint(1, 50)

limite_tentativas = 10
tentativas = 0

while True:
    if tentativas >= limite_tentativas:
        print(f"Você chegou ao limite de {limite_tentativas} tentativas.")
        print(f"O número era {numero_secreto}.")
        break

    palpite = int(input(f"Tentativa {tentativas + 1}/{limite_tentativas}: "))

    tentativas += 1

    if palpite == numero_secreto:
        print(f"Parabéns! Você acertou em {tentativas} tentativas.")
        break
    elif palpite < numero_secreto:
        print("Seu palpite é menor que o número. Tente um número maior.")
    else:
        print("Seu palpite é maior que o número. Tente um número menor.")
```
